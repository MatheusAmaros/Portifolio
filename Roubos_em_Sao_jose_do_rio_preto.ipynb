# 2. Operações básicas entre geometrias

```python
#vamos crias dois circulos
circulo_1 = Point(1,1).buffer(1)
circulo_2 = Point(2,1).buffer(1)
circulo_2
```

```python
import matplotlib.pyplot as plt
```

```python
x,y = circulo_1.exterior.xy
plt.plot(x,y)

x,y = circulo_2.exterior.xy
plt.plot(x,y)

plt.show()
```

```python
#obtendo a interseção do circulo
circulo_inter = circulo_1.intersection(circulo_2)
circulo_inter
```

```python
#obtendo a uniao dos circulos
uniao = circulo_1.union(circulo_2)
uniao
```

```python
#diferença dos circulos
dif_1 = circulo_1.difference(circulo_2)
dif_1
```

```python
#diferença simetrica
dif_sim = circulo_1.symmetric_difference(circulo_2)
dif_sim
```

# Criar o GeoDataFrame(geopandas)

```python
from shapely.geometry import Point
```

```python
#criar a coluna geometry
df_roubos['geometry'] = None

for index, row in df_roubos.iterrows():
    df_roubos.loc[index,'geometry'] = Point(row.LONGITUDE, row.LATITUDE)
```

```python
df_roubos.head(50)
```

```python
#criar o geodataframe
gdf_roubos = gpd.GeoDataFrame(df_roubos, geometry='geometry')
```
